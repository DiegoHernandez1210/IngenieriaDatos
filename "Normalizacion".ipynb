##Normalización

```python
import pandas as pd

# 1. Cargar archivo
df_compact = pd.read_csv("WRegularSeasonCompactResults.csv")

# 2. Crear tabla de equipos (sin duplicados)
teams_compact = pd.DataFrame(
    pd.concat([df_compact['WTeamID'], df_compact['LTeamID']]).unique(),
    columns=['TeamID']
)

# 3. Crear tabla de partidos (estructura limpia)
games_compact = df_compact[['Season', 'DayNum', 'WTeamID', 'LTeamID', 'WScore', 'LScore', 'WLoc', 'NumOT']]

# 4. Renombrar columnas para mejor entendimiento
games_compact = games_compact.rename(columns={
    'WTeamID': 'WinnerTeamID',
    'LTeamID': 'LoserTeamID',
    'WScore': 'WinnerScore',
    'LScore': 'LoserScore'
})

# 5. Mostrar resultados
print("Tabla Equipos:")
print(teams_compact.head())

print("\nTabla Partidos:")
print(games_compact.head())
```

##Desnormalización

```python
import pandas as pd

# 1. Cargar archivo
df_compact = pd.read_csv("WRegularSeasonCompactResults.csv")

# 2. Crear tabla de equipos
teams_compact = pd.DataFrame(
    pd.concat([df_compact['WTeamID'], df_compact['LTeamID']]).unique(),
    columns=['TeamID']
)

# 3. Crear nombres simulados de equipos
teams_compact['TeamName'] = "Equipo_" + teams_compact['TeamID'].astype(str)

# 4. Unir datos del ganador
df_denorm_compact = df_compact.merge(
    teams_compact, left_on='WTeamID', right_on='TeamID', how='left'
).rename(columns={'TeamName': 'WinnerName'})

# 5. Unir datos del perdedor
df_denorm_compact = df_denorm_compact.merge(
    teams_compact, left_on='LTeamID', right_on='TeamID', how='left'
).rename(columns={'TeamName': 'LoserName'})

# 6. Eliminar columnas duplicadas innecesarias
df_denorm_compact = df_denorm_compact.drop(columns=['TeamID_x', 'TeamID_y'])

# 7. Mostrar resultado
print("Tabla desnormalizada:")
print(df_denorm_compact.head())
```

```python
import pandas as pd

# 1. Cargar archivo
df_slots = pd.read_csv("WNCAATourneySlots.csv")

# 2. Crear tabla de equipos (semillas)
teams_slots = pd.DataFrame(
    pd.concat([df_slots['StrongSeed'], df_slots['WeakSeed']]).unique(),
    columns=['Seed']
)

# 3. Crear nombres simulados de equipos
teams_slots['TeamName'] = "Equipo_" + teams_slots['Seed'].astype(str)

# 4. Unir datos del equipo fuerte
df_denorm_slots = df_slots.merge(
    teams_slots, left_on='StrongSeed', right_on='Seed', how='left'
).rename(columns={'TeamName': 'StrongTeamName'})

# 5. Unir datos del equipo débil
df_denorm_slots = df_denorm_slots.merge(
    teams_slots, left_on='WeakSeed', right_on='Seed', how='left'
).rename(columns={'TeamName': 'WeakTeamName'})

# 6. Eliminar columnas duplicadas innecesarias
df_denorm_slots = df_denorm_slots.drop(columns=['Seed_x', 'Seed_y'])

# 7. Mostrar resultado
print("Tabla desnormalizada:")
print(df_denorm_slots.head())
```

```python
import pandas as pd

# 1. Cargar archivo
df_seeds = pd.read_csv("WNCAATourneySeeds.csv")

# 2. Visualizar estructura
print(df_seeds.head())

# 3. Crear tabla de temporadas (sin duplicados)
seasons = df_seeds[['Season']].drop_duplicates()

# 4. Crear tabla de equipos (IDs únicos)
teams = df_seeds[['TeamID']].drop_duplicates()

# 5. Crear tabla de semillas (estructura principal)
seeds = df_seeds[['Season', 'Seed', 'TeamID']]

# 6. Separar tipo de seed (opcional - mejora de normalización)
seeds['SeedRegion'] = seeds['Seed'].str[0]   # Ej: W, X, Y, Z
seeds['SeedNumber'] = seeds['Seed'].str[1:]  # Ej: 01, 02, etc.

# 7. Mostrar resultados
print("Tabla Seasons:")
print(seasons.head())

print("\nTabla Teams:")
print(teams.head())

print("\nTabla Seeds:")
print(seeds.head())
```

##Desnormalización

```python
import pandas as pd

# 1. Cargar archivo
df_seeds = pd.read_csv("WNCAATourneySeeds.csv")

# 2. Crear tabla de equipos
teams = df_seeds[['TeamID']].drop_duplicates()

# 3. Crear nombres simulados
teams['TeamName'] = "Equipo_" + teams['TeamID'].astype(str)

# 4. Unir datos de equipos
df_denorm_seeds = df_seeds.merge(
    teams, on='TeamID', how='left'
)

# 5. Separar información de la seed (opcional)
df_denorm_seeds['Region'] = df_denorm_seeds['Seed'].str[0]
df_denorm_seeds['NumeroSeed'] = df_denorm_seeds['Seed'].str[1:]

# 6. Mostrar resultado
print("Tabla desnormalizada:")
print(df_denorm_seeds.head())
```
